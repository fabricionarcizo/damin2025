# Logistic Regression Classification with scikit-learn

This notebook demonstrates how to use the Logistic Regression algorithm to classify data points using the scikit-learn library. It shows the sigmoid curve, decision boundary, and probability predictions for binary classification.

## Import Required Libraries

Import the necessary libraries for data generation, modeling, and visualization.

```python
# Import the required libraries.
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression

from sklearn.datasets import make_classification
```

## Generate Binary Classification Dataset

Create a synthetic 1D binary classification dataset using scikit-learn's make_classification function.

```python
# Generate a 1D binary classification dataset.
X, y = make_classification(
    n_samples=100, n_features=1, n_informative=1, n_redundant=0,
    n_clusters_per_class=1, random_state=42
)
```

## Train Logistic Regression Model

Initialize and train a Logistic Regression model on the generated dataset.

```python
# Initialize and train the Logistic Regression model.
model = LogisticRegression()
model.fit(X, y)
```

## Predict Probabilities

Generate predictions across a range of feature values to create the sigmoid curve.

```python
# Predict probabilities for the logistic curve.
X_test = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]
```

## Visualize Logistic Regression Results

Plot the dataset points, sigmoid curve, and decision boundary to illustrate how Logistic Regression works.

```python
# Plotting the dataset and logistic curve.
plt.figure(figsize=(10, 6))

# Scatter plot of the dataset.
plt.scatter(X[y == 0], y[y == 0], color="#2580B7", marker="o", label="Class 0")
plt.scatter(X[y == 1], y[y == 1], color="#179E86", marker="x", label="Class 1")

# Plot logistic sigmoid curve.
plt.plot(X_test, y_prob, color="#C03B26", label="Sigmoid Curve")

# Plot the decision boundary.
plt.axvline(
    x=model.intercept_ / -model.coef_, color="#44546A", linestyle="--",
    label="Decision Boundary"
)

# Labels and legend.
plt.title("Logistic Regression with Sigmoid Curve")
plt.xlabel("Feature 1")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```
