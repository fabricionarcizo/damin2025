# Binary Classification Confusion Matrix

This notebook demonstrates how to create and visualize a confusion matrix for binary classification using scikit-learn and seaborn. A confusion matrix helps evaluate classification model performance by showing the counts of true positives, true negatives, false positives, and false negatives.

## Import Required Libraries

Import libraries for data generation, modeling, evaluation, and visualization.

```python
# Import the required libraries.
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.colors import LinearSegmentedColormap
from sklearn.tree import DecisionTreeClassifier

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

## Step 1: Generate Binary Classification Dataset

Create a synthetic dataset with 5 features for binary classification using scikit-learn's make_classification function.

```python
# Step 1: Generate a dataset for binary classification.
X, y = make_classification(
    n_samples=1000, n_features=5, n_informative=3,
    n_redundant=1, n_classes=2, random_state=42
)
```

## Step 2: Split Data into Training and Testing Sets

Divide the dataset into training and testing subsets for model evaluation.

```python
# Step 2: Split the data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## Step 3: Train Decision Tree Classifier

Initialize and train a Decision Tree classifier for binary classification.

```python
# Step 3: Train a Decision Tree Classifier for binary classification.
classifier = DecisionTreeClassifier(random_state=42)
classifier.fit(X_train, y_train)
```

## Step 4: Make Predictions

Use the trained model to predict classes for the test dataset.

```python
# Step 4: Make predictions on the test set.
y_pred = classifier.predict(X_test)
```

## Step 5: Compute Confusion Matrix

Calculate the confusion matrix to evaluate classification performance.

```python
# Step 5: Compute the confusion matrix.
cm = confusion_matrix(y_test, y_pred)
```

## Step 6: Visualize Confusion Matrix

Plot the confusion matrix as a heatmap with annotations for True Positives (TP), False Negatives (FN), False Positives (FP), and True Negatives (TN).

```python
# Step 6: Plot the confusion matrix with TP, FN, FP, TN.
plt.figure(figsize=(8, 6))
cmap = LinearSegmentedColormap.from_list(
    "custom_green", ["#FFFFFF", "#179E86"])
ax = sns.heatmap(
    cm, annot=True, fmt="d", cmap=cmap,
    xticklabels=["Predicted Positive", "Predicted Negative"],
    yticklabels=["Actual Positive", "Actual Negative"]
)

# Annotate the confusion matrix with TP, TN, FP, FN.
ax.text(
    0.2, 0.2, "TP", horizontalalignment="center",
    verticalalignment="center", color="white", fontsize=14
)
ax.text(
    0.2, 1.8, "FP", horizontalalignment="center",
    verticalalignment="center", color="black", fontsize=14
)
ax.text(
    1.8, 0.2, "FN", horizontalalignment="center",
    verticalalignment="center", color="black", fontsize=14
)
ax.text(
    1.8, 1.8, "TN", horizontalalignment="center",
    verticalalignment="center", color="white", fontsize=14
)

plt.title("Confusion Matrix with TP, FN, FP, and TN")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

## Understanding Confusion Matrix Components

- **True Positive (TP)**: Correctly predicted positive cases
- **True Negative (TN)**: Correctly predicted negative cases  
- **False Positive (FP)**: Incorrectly predicted positive cases (Type I error)
- **False Negative (FN)**: Incorrectly predicted negative cases (Type II error)

The confusion matrix provides a comprehensive view of classification performance and is the foundation for calculating metrics like precision, recall, and F1-score.
